## Predictive Modeling

This notebook builds a model to predict whether a client will subscribe 
to a term deposit, based on the cleaned and preprocessed dataset.

**Note on `duration`:** although duration is one of the strongest predictor in the dataset, 
I have excluded it from the model, whose purpose is to answer this question: "Is it worth calling or contacting this specific customer?".
Including this variable would create a paradox: at the time of prediction, the call has not yet taken palces, making that data unavailable in a real-world scenario.
We know that longer calls (after a certain duration) have diminishing return anyway.

The goal is a model that is not just accurate, but actually realistic.

## 2. Data Preprocessing

## Encoding strategy

The encoding strategy was chosen based on the nature of each feature and the type of information it represents.

All categorical variables in the dataset are **nominal features** (no natural order - although `month` and `day_of_week` have an order, this is a temporal order so they will be treated as nominal categorical variables to not introduce misleading numerical relationships between thse categories).  

Because of this, **One-Hot Encoding** was selected as the main encoding technique.

 Only for the target (binary), I will use **manual mapping** to encode it.


## 2.1 Encoding for target

```python
## Encoding for target
y_train_encoded= y_train.map({
    'no': 0,
    'yes': 1})

y_test_encoded= y_test.map({
    'no': 0,
    'yes': 1})

# I named these with 'encoded' attribute to mantain consistency with the name given to the feature ('X_train_encoded' and 'X_test_encoded')
```

## 3. Baseline Model Training, Evaluation, Comparison

## Handling Class Imbalance

The dataset is heavily imbalanced, roughly 88% of clients did not subscribe. Without addressing this, any model would learn 
to simply predict "no" for everyone and still achieve 88% accuracy, 
which is completely useless in practice.

Rather than synthetically generating new data points (SMOTE) or removing 
majority class samples (both of which alter the dataset)
I chose to work with the data as it is and inform the models directly 
about the class distribution.

For Logistic Regression and Random Forest, I use `class_weight='balanced'`, 
which automatically adjusts the penalty for misclassifying the minority class. 
For XGBoost, the equivalent parameter is `scale_pos_weight`, set to the ratio 
of negative to positive samples - telling the model how much more attention 
to pay to the minority class during training.

This approach for me is cleaner and avoids introducing 
artificial bias into the training data.

## 3.1 Models - Fit and Score

Rather than jumping straight to the most complex model, I will train and compare 
three algorithms of increasing complexity. This approach is more rigorous because
if a simpler model performs comparably to a complex one, there is little reason 
to add the extra complexity.

**1. Logistic Regression**: serves as baseline, it is the simplest possible 
classifier and sets the floor. Any model worth using should comfortably 
outperform it. I will set `class_weight='balanced'` to better handle the imbalanced dataset.

**2.Random Forest**: adds complexity through ensemble learning, useful to capture non-linear relationships that logistic regression 
cannot model (`class_weight='balanced'`).

**XGBoost** is the primary candidate beacause it handles class imbalance natively through 
`scale_pos_weight`, and consistently outperforms other algorithms on 
structured datasets with mixed feature types.

The final model selection will be based on ROC-AUC and F1-score as the primary metric.
Accuracy alone is misleading on imbalanced datasets where predicting 
"no" for everyone would already yield 88% accuracy in this case.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost
from xgboost import XGBClassifier
```

```python
print(xgboost.__version__)
```

```python
# Calculate scale_pos_weight for XGB
# Following the XGB documentation, this is the ratio of negative instances to positive instances in the training data
negative= (y_train_encoded==0).sum()
positive= (y_train_encoded==1).sum()

scale_pos_weight= negative/positive
print(f"Scale_pos_weight: {scale_pos_weight:.2f}")
```

```python
# Put models in a dictionary
models={
    "Logistic Regression": LogisticRegression(class_weight='balanced'),
    "Random Forest": RandomForestClassifier(class_weight='balanced'),
    "XGBoost Classifier": XGBClassifier(scale_pos_weight=scale_pos_weight)}

# Function to fit and score models
def fit_score(models, X_train_encoded, X_test_encoded, y_train_encoded, y_test_encoded):
    """
    Fits and evaluates given ML models 
    using ROC-AUC and F1-score.

    Parameters:
    models: a dict of differetn ML models
    X_train_encoded: encoded training data
    X_test_encoded: encoded testing data
    y_train_encoded: encoded training labels
    y_test_encoded: encoded test labels

    Output:
    dataFrame with model performance metrics.
    """
    np.random.seed(42)
    
    # A dictionary to keep the model scores
    model_scores={}

    # Loop through the models
    for name, model in models.items():
        # Fit the model
        model.fit(X_train_encoded, y_train_encoded)
        # Predictions
        y_pred=model.predict(X_test_encoded)
        # Probabilities for ROC-AUC
        y_probs=model.predict_proba(X_test_encoded)[:,1]
        # Calculate metrics
        roc_auc=roc_auc_score(y_test_encoded, y_probs)
        f1=f1_score(y_test_encoded, y_pred)
        # Append the score to model_scores
        model_scores[name]={
            'ROC-AUC': roc_auc,
            'F1-score': f1}
    
    return pd.DataFrame(model_scores).T
```

```python
model_scores= fit_score(models= models,
                        X_train_encoded = X_train_encoded,
                        X_test_encoded= X_test_encoded,
                        y_train_encoded= y_train_encoded,
                        y_test_encoded= y_test_encoded)
model_scores
```

## NOTE:
**Logistic Regression initially did not converge within the default number of iterations (max_iter=100).**
This is common in datasets with mixed feature scales and after one-hot encoding (it increases dimensionality).

To address this:
1. I will increase max_iter to ensure the optimization algorithm has enough iterations to converge.
2. I will include StandardScaler to improve optimization stability ( using a Pipeline to structure preprocessing and modeling in a clean way).

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
log_reg= Pipeline([
        ('scaler', StandardScaler(with_mean=False)), #with_mean=False because One-Hot Encoding create a sparse matrix with a lot of 0
        ('model', LogisticRegression(max_iter=1000, class_weight='balanced'))])
```

```python
models_2={
    "Logistic Regression": log_reg,
     "Random Forest": RandomForestClassifier(class_weight='balanced'),
    "XGBoost Classifier": XGBClassifier(scale_pos_weight=scale_pos_weight)}

model_scores= fit_score(models= models_2,
                        X_train_encoded = X_train_encoded,
                        X_test_encoded= X_test_encoded,
                        y_train_encoded= y_train_encoded,
                        y_test_encoded= y_test_encoded)
model_scores
```

## 3.2 Cross-Validation
To obtain a more reliable comparison between models, cross-validation is performed after the initial train/test evaluation.

Since model performance can vary depending on a single split, I will use **Stratified K-Fold** cross-validation as it provides a more robust estimate by evaluating each model across multiple folds while preserving class distribution in each of them (this is particularly important for an imbalanced dataset).

```python
from sklearn.model_selection import cross_validate, StratifiedKFold
```

```python
cv= StratifiedKFold(n_splits=5,
                    shuffle=True,
                    random_state=42)

scoring=['roc_auc', 'f1']

# A dictionary to keep the scores
cross_val_scores={}

for name, model in models_2.items():
    scores= cross_validate(
        model,
        X_train_encoded,
        y_train_encoded,
        cv=cv,
        scoring=scoring)
    
    cross_val_scores[name]={
        'Mean ROC-AUC': scores['test_roc_auc'].mean(),  # mean -> synthetic value
        'Mean F1-score': scores['test_f1'].mean()}
    
```

```python
cross_val_df= pd.DataFrame(cross_val_scores).T
cross_val_df
```

## 4. Hyperparameter Tuning

## 4.1 Model Selection for tuning
Cross-validation results show **Logistic Regression** outperforming both 
Random Forest and XGBoost by approximately 2 points in ROC-AUC (0.788 vs ~0.765). 
This is not entirely surprising, the EDA revealed that some of the strongest predictors 
in this dataset (euribor, duration) have a largely linear relationship with 
the target, and maybe this is why Logistic Regression has this strength in this case.

Random Forest underperforms on F1-score (0.36), it struggles 
more with the class imbalance .

For these reasons, I will tune **Logistic Regression** with RandomizedSearchCV because it achieved the best balance between ROC-AUC and F1-score while also being more interpretable and simpler (computanionally) to the ensemble-based approaches.

But I will also tune **XGBoost** (RandomizedSearchCV) - it has significantly more levers to pull than 
Logistic Regression, and the default parameters are rarely optimal. 
If tuned XGBoost then exceeds Logistic Regression, I'll select it 
for its superior ability to capture non-linear interactions and I will try further improvements with GridSearchCV:
If not, Logistic Regression wins (it's simpler, faster, and more interpretable) and I will continue with it.

## 4.2 RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

I selected for logistic regression:
- 'lbfg' and 'saga'-> they handle medium/high dimensional datasets and ensure stable convergence.
- 'l2'->  it helps prevent overfitting by shrinking coefficients without forcing sparsity (mantain all features).

```python
# Hyperparameter grid for logistic Regression
log_reg_grid={
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['lbfgs', 'saga'],
    'penalty': ['l2', None]}

# Hyperparameter grid for logistic Regression
XGB_grid= {
    'max_depth': [3, 4, 5, 6],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.6, 0.8, 1.0],
    'scale_pos_weight': [scale_pos_weight]}
```

**Note**: below -> only 'roc_auc' to evaluate models because it's more robust on imbalanced dataset and unlinke F1 (which depends on a specific classification threshold - default 0.5), it measures the model's ability to distinguish between classes across all possible thresholds.

The selected model, after this tuning, will be optimized using threshold tuning based on F1-score and recall.

```python
# Tune Logistic Regression
np.random.seed(42)

log_reg_tuned= RandomizedSearchCV(
    LogisticRegression(class_weight='balanced', max_iter=1000),
    param_distributions= log_reg_grid,
    n_iter=20,
    scoring='roc_auc',
    cv=5,
    n_jobs=-1)

log_reg_tuned.fit(X_train_encoded, y_train_encoded)

print(f"Best params: {log_reg_tuned.best_params_}")
```

```python
# Evaluate the randomized search Logistic Regression model
y_pred_log_reg_tuned= log_reg_tuned.predict(X_test_encoded)
y_prob_log_reg_tuned= log_reg_tuned.predict_proba(X_test_encoded)[:,1]

print("Logistic Regression Tuned")
print(f"ROC-AUC: {roc_auc_score(y_test_encoded, y_prob_log_reg_tuned)}")
```

```python
# Tune XGB Classifier
np.random.seed(42)

XGB_tuned= RandomizedSearchCV(
    XGBClassifier(eval_metric='auc'),
    param_distributions= XGB_grid,
    n_iter=40,
    scoring='roc_auc',
    cv=5,
    n_jobs=-1)

XGB_tuned.fit(X_train_encoded, y_train_encoded)

print(f"Best params: {XGB_tuned.best_params_}")
    
```

```python
# Evaluate the randomized search XGB model on X_test, y_test
y_pred_xgb_tuned= XGB_tuned.predict(X_test_encoded)
y_prob_xgb_tuned= XGB_tuned.predict_proba(X_test_encoded)[:,1]

print("XGB Tuned")
print(f"ROC-AUC: {roc_auc_score(y_test_encoded, y_prob_xgb_tuned)}")
```

## 4.3 Visual Comparison

```python
roc_auc_score_lr_tuned= roc_auc_score(y_test_encoded, y_prob_log_reg_tuned)
roc_auc_score_xgb_tuned= roc_auc_score(y_test_encoded, y_prob_xgb_tuned)

roc_auc_results=pd.DataFrame({
    "Model": ['XGBoost Tuned', 'Log-Reg Tuned'],
    "ROC-AUC":[roc_auc_score_xgb_tuned, roc_auc_score_lr_tuned]})

roc_auc_results.plot.bar(x='Model', y='ROC-AUC', legend= False)
plt.title('Model Comparison - ROC-AUC')
plt.ylabel('ROC-AUC')
plt.xticks(rotation= 45)
plt.show()
```

```python
sns.set(font_scale= 1.5)

# Write a function to plot confusion matrix using Seaborn heatmap()
def plot_conf_matrix(y_test_encoded, y_pred_xgb_tuned):
    fig, ax= plt.subplots(figsize=(6, 5))
    ax= sns.heatmap(confusion_matrix(y_test_encoded, y_pred_xgb_tuned),
                   annot= True,
                   cbar= False,
                   fmt= 'd')
    
    plt.title( 'Confusion Matrix')
    plt.xlabel('True label')
    plt.ylabel('Predicted label')

    bottom, top= ax.get_ylim()
    ax.set_ylim()
```

```python
plot_conf_matrix(y_test_encoded, y_pred_xgb_tuned)
```

The tuned XGBoost model shows obviously strong overall performance in identifying non-churn customers (class 0), with high precision and recall for the majority class.

For churn prediction (class 1), the model achieves a recall of 0.66, meaning it successfully identifies 66% 'yes' (subscription). However, precision remains relatively low (0.40), indicating the presence of false positives.

The confusion matrix confirms this trade-off:
- 626 customers were correctly identified (True Positives)
- 318 customers were missed (False Negatives)

This model provides a reasonable baseline before threshold optimization, considering that the best goal fo this specific case is to minimize missed potential subscribers (capturing more real 'yes').

## 6. Threshold Optimization
By default, classification models use a decision thresholf of 0.5 to convert predicted probabilities into class labels. However, this default threshold is not optimal in an imbalanced dataset and classification problem. Therefore, relying only on the default threshold could lead to missing valuable churn cases (False Negatives).

To improve the decision-making process -> threshold optimization using 2 different approaches:
1. **F1-score threshold optimization** - maximize the F1-score in order to balance precision and recall (a general trade-off between correctly identifying customers/subscribers while limiting false positives).
2. **Recall threshold optimization** - optimization strategy focused on maximizing recall. In a business context like this, identifying more potential subscribers is often more important than maintaining high precision, as failing to detect a potential subscriber may lead to direct customer loss. Obviously, I will do this while maintaining a baseline level of precision.

**Note**: cost of pursuing the wrong leads (beacuse higher recall and less precision -> more false positives) will be ignored here (only subscription vs no subscription -> 'to call or not to call?').

## 6.1 Optimal Threshold - F1 score

```python
from sklearn.metrics import precision_recall_curve
```

```python
# Precision, recall for each threshold
precisions, recalls, thresholds= precision_recall_curve(y_test_encoded, y_prob_xgb_tuned)
precisions, recalls, thresholds
```

```python
# Calculate F1 for each thresholds
f1_scores= 2 * (precisions[:-1] * recalls[:-1]) / (precisions[:-1] + recalls[:-1])
f1_scores
```

```python
# Find the best threshold
best_index= np.argmax(f1_scores)
best_threshold= thresholds[best_index]

print(f"Best Threshold: {best_threshold}")
```

## Evaluation

```python
best_y_pred= (y_prob_xgb_tuned >= 0.5793).astype(int)

# Classification report with the best threshold
print("XGBoost Tuned - Best Threshold (0.5793)")
print(classification_report(y_test_encoded, best_y_pred))
```

```python
# Confusion matrix with the best threshold
print(confusion_matrix(y_test_encoded, best_y_pred))
# Visualize it
plot_conf_matrix(y_test_encoded, best_y_pred)
```

 
Compared to the default threshold, after optimizing the threshold based on the F1-score, the model achieves a better balance between precision and recall (as can be seen in the classification report).

The confusion matrix shows a reduction in false positives (improving overall classification stability), but also an increase in the number of False Negatives (359 missed potential subscriber here vs 318).

Fo this reason, I will try to optimize the threshold focusing on recall.

## 6.2 Optimal Threshold - Recall (with a precision >= 0.35)
Prioritizing the identification of 'yes', but with a precision constraint of at least 0.35 to avoid an excessive number of false positives.

```python
# Precision, recall for each threshold
precisions, recalls, thresholds= precision_recall_curve(y_test_encoded, y_prob_xgb_tuned)

# Keep only thresholds where precision is at least 0.35
indexes= np.where(precisions[:-1] >= 0.35)[0]

# Among these thresholds take the one with the highest recall
best_index= indexes[np.argmax(recalls[:-1][indexes])]

# Calculate the best threshold
best_threshold_rec= thresholds[best_index]

best_recall= recalls[best_index]

print(f"Best Thresholds: {best_threshold}")
print(f"Recall: {best_recall}")
```

```python
# Final prediction
best_y_pred_2= (y_prob_xgb_tuned >= best_threshold_rec).astype(int)
# Classification report with the best threshold
print(classification_report(y_test_encoded, best_y_pred_2))
```

```python
# Confusion matrix with the best threshold
print(confusion_matrix(y_test_encoded, best_y_pred_2))
# Visualize it
plot_conf_matrix(y_test_encoded, best_y_pred_2)
```

## 6.3 Threshold Comparison and Business Trade-Offs 
The recall-optimized threshold increases the model’s ability to identify 'yes', achieving a recall of 0.71 for the positive class.

Compared to the previous thresholds, the number of True Positives increases from 585 to 669, while False Negatives decrease from 359 to 275. Although this comes at the cost of more False Positives and lower overall precision, this trade-off is acceptable in a business context where missing potential customers/subscribers is more costly than incorrectly flagging (and call) non-customers/non-subscribers.

**Decision**: XGB tuned with the recall-optimized threshold (0.4243335723876953).
